This notebook shows the process of making a pair-plot in Matplotlib.  You can do it, but it's a lot of boring work.  Seaborn has wrappers over Matplotlib that make this process simpler - at the expense of some loss of generality in how easily you can customize the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('premier_league_2021.csv')
df = df[['wages_year', 'goal_difference']]
df.head()
```

```python
# A pair plot using Matplotlib.
fig, axes = plt.subplots(2, 2)
axes[0, 0].hist(df['wages_year'])
axes[0, 0].set_ylabel('wages_year')
axes[0, 0].set_xticks([])
axes[1, 0].scatter(df['wages_year'], df['goal_difference'])
axes[1, 0].set_xlabel('wages_year')
axes[1, 0].set_ylabel('goal_difference')
axes[1, 0].set_xticks([])
axes[0, 1].scatter(df['goal_difference'], df['wages_year'])
axes[1, 1].hist(df['goal_difference'])
axes[1, 1].set_xlabel('goal_difference');
```

```python
# A pair plot with Seaborn.
sns.pairplot(df)
```
